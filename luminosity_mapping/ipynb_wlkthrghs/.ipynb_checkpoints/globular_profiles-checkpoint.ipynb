```python
import numpy as np
import os
import yt
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns 
import matplotlib.cm as cm
import plotly.graph_objects as go

from lum_funcs import look_up_table, get_cluster
```

```python
pop_2_data = np.loadtxt(r"pop_2_data/pos_00500_479_16_myr.txt")  

birth_epochs = pop_2_data[:,0] *1e6   #yr
ages = pop_2_data[:,1] *1e6           #yr
current_time = pop_2_data[0,6]        #yr

ages [ages < 1e6 ] = 1e6 #if the age <10^6 yr, set the age to 10^6 year.
x_pos = pop_2_data[:,2]
y_pos = pop_2_data[:,3] 
z_pos = pop_2_data[:,4]
masses = pop_2_data[:,5] #Msun

stellar_luminosities = look_up_table(
    stellar_ages=ages, 
    table_link='https://www.stsci.edu/science/starburst99/data/l1500_inst_e.dat', 
    column_idx=1,
    log=True)
scaled_stellar_lums = stellar_luminosities*1e-5
```

Initial analysis.

```python
plt.figure(figsize = (16,8), )
plt.subplot(1, 2, 1)
plt.scatter(ages, scaled_stellar_lums , label= 'Pop II Data', alpha=0.5)
plt.title(r"Assigned Monochromatic luminosity at 1500 Å using $\alpha=2.35$")
plt.xlabel("Stellar Age (yr)")
plt.ylabel("Luminosity (erg/s/Å)")
plt.xscale('log')
plt.yscale('log')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(scaled_stellar_lums, 
         bins = 10, 
         label = "(Min, Max, Mean) = ({:.1e}, {:.1e}, {:.1e}) erg/s/Å \nCounts = {} Stars".format(
             np.min(scaled_stellar_lums ), np.max(scaled_stellar_lums ), np.mean(scaled_stellar_lums), len(ages)),
         color='tab:red'
        )
plt.xlabel('Pop II Luminosities')
plt.ylabel('Counts')
plt.title('Simulation Pop II Luminosity Distribution at t = {:.3f}Myr'.format(current_time))
plt.yscale('log')
plt.legend()
plt.show()
```

```python
recentered_x = sph_x - np.median(sph_x)
recentered_y = sph_y - np.median(sph_y)
recentered_z = sph_z - np.median(sph_z)

recentered = go.Scatter3d(
    x=recentered_x, 
    y=recentered_y, 
    z=recentered_z, 
    opacity=0.1, 
    mode='markers',
    marker=dict(color=np.log10(sph_lums), 
                showscale=True, 
                size=1,
                colorscale='inferno',
                colorbar=dict(title='Log Monochromatic luminosity at 1500 Å'), 
               ),
    )
recentered.marker.colorbar.title.side = 'right'

fig=go.Figure()
fig.add_trace(recentered)

fig.update_layout(title = 't = {:.3f}Myr'.format(current_time), 
                  template='plotly_dark', 
                  width=700, 
                  height=600,
                  scene = dict(
                  xaxis = dict(nticks=3, ),
                  yaxis = dict(nticks=3, showgrid=False),
                  zaxis = dict(nticks=3, showgrid=False),
                  ),
                 )
fig.update_scenes(xaxis_title=dict(text='x (pc)'),
                  yaxis_title=dict(text='y (pc)'),
                  zaxis_title=dict(text='z (pc)'),
                  )
fig.show()
```

```python
def shell_mask(xpos, ypos, zpos, ctr_at, outer_radius, lums, dr):
    all_positions = np.vstack((xpos, ypos, zpos)).T 
    distances = np.sqrt(np.sum(np.square(all_positions - ctr_at), axis=1)) 
    mask = ((outer_radius - dr) <= distances) & (distances <= outer_radius)
    #print(distances.size)
    masked_positions = all_positions[mask]
    masked_lums = lums[mask]
    return masked_positions[:,0], masked_positions[:,1], masked_positions[:,2], masked_lums

def surface_brightness(dr):
    d_r = np.arange(dr,10+dr,dr)
    surface_lums = []
    counts = []
    for i,outer_r in enumerate(d_r):
        x, y, z, lums =  shell_mask(
            recentered_x, recentered_y, recentered_z, np.array([0,0,0]), outer_radius=outer_r, lums=sph_lums, dr=dr 
            )
        
        surface_lums.append(np.sum(lums)) 
        counts.append(len(lums))
        #plt.figure(figsize = (8,8), )
        plt.scatter(x,y,s=1)
        plt.xlim(-20,20)
        plt.ylim(-20,20)
        plt.savefig('test_sequence/dr_05/{:05d}.png'.format(i))
        plt.clf() 
        
    return d_r, np.array(surface_lums), np.array(counts)
```

```python
def ring_2d_mask(xpos, ypos, ctr_at, outer_radius, lums, dr):
    all_positions = np.vstack((xpos, ypos)).T 
    distances = np.sqrt(np.sum(np.square(all_positions - ctr_at), axis=1)) 
    mask = ((outer_radius - dr) <= distances) & (distances <= outer_radius)
    #print(distances.size)
    masked_positions = all_positions[mask]
    masked_lums = lums[mask]
    return masked_positions[:,0], masked_positions[:,1], masked_lums

def surface_2d_brightness(dr):
    d_r = np.arange(dr,10+dr,dr)
    surface_lums = []
    counts = []
    for i,outer_r in enumerate(d_r):
        x, y, lums =  ring_2d_mask(
            recentered_x, recentered_y, np.array([0,0]), outer_radius=outer_r, lums=sph_lums, dr=dr 
            )
        
        surface_lums.append(np.sum(lums)) 
        counts.append(len(lums))
        #plt.figure(figsize = (8,8), )
        plt.scatter(x,y,s=1)
        plt.xlim(-20,20)
        plt.ylim(-20,20)
        plt.savefig('test_sequence/dr_05_flat/{:05d}.png'.format(i))
        #print(i)
        plt.clf() 
        
    return d_r, np.array(surface_lums), np.array(counts)
```
